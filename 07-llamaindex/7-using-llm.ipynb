```python
from pydantic import BaseModel
from llama_index.core.tools import FunctionTool

class Song(BaseModel):
    """A song with name and artist"""
    name: str
    artist: str
    

def generate_song(name: str, artist: str) -> Song:
    """Generates a song with provided name and artist"""
    
    return Song(name=name, artist=artist)


tools = FunctionTool.from_defaults(fn=generate_song)

response = llm.predict_and_call(
                    tools=[tools], 
                    user_msg="Pick a random telugu song for me")

response
```

```python
response.response
```

Manual Tool Call Invocation

```python
chat_history = [ChatMessage(role="user", content="Pick 2 random songs for me")]

response = llm.chat_with_tools([tools], chat_history=chat_history)
```

```python
response.message.additional_kwargs
```

```python

tool_calls = llm.get_tool_calls_from_response(
    response, error_on_no_tool_call=False #Supress error when there are no tool call
)

tool_calls
```

```python
while tool_calls:
    chat_history.append(response.message)
    for tool_call in tool_calls:
        tool_name = tool_call.tool_name
        tool_kwargs = tool_call.tool_kwargs
        print(f"Calling {tool_name} with {tool_kwargs}")
        
        tool_output = tools(**tool_kwargs)
        
        chat_history.append(
            ChatMessage(role="tool", content=str(tool_output),
                        additional_kwargs={"tool_call_id": tool_call.tool_id})
        )
        
        response = llm.chat_with_tools([tools], chat_history=chat_history)
        tool_calls = llm.get_tool_calls_from_response(
            response, error_on_no_tool_call=False
        )
        
```

```python
from typing import List, Optional
from pydantic import BaseModel

class Actor(BaseModel):
    """An actor in a movie"""
    name: str
    character_name: str
    is_lead_role: bool
    

class Movie(BaseModel):
    """A movie with title, director, genre and cast"""
    title: str
    director: str
    release_year: int
    genres: List[str]
    rating: float
    cast: List[Actor]
    synopsis: Optional[str] = None
```

```python
from llama_index.core.prompts import PromptTemplate

prompt_template = PromptTemplate(
    "Recommend a popular {genre} telugu movie with details"
)

response = llm.as_structured_llm(Movie).complete(prompt_template.format(genre="sci-fi"))

response
```

```python
movie_obj = llm.structured_predict(Movie, prompt_template, genre="action")

movie_obj
```
