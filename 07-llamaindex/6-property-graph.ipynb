Persisting Index to Chromadb VectorStore

```python
pip install llama-index-vector-stores-chroma
```

```python
import chromadb
from llama_index.vector_stores.chroma import ChromaVectorStore
from llama_index.core.graph_stores import SimplePropertyGraphStore

client = chromadb.PersistentClient("./chroma_db")

# Here Collection is like a database schema
collection = client.get_or_create_collection("my_graph_vector_db")
vector_store = ChromaVectorStore(
    chroma_collection=collection
)

index = PropertyGraphIndex.from_documents(
    documents=documents,
    vector_store=vector_store, #Stored in Persisted Chroma store
    property_graph_store=SimplePropertyGraphStore(), #Inmemory Graph Store
    show_progress=True
)

index.storage_context.persist(persist_dir="./storage4")
```

Loading Index from Chroma VectorDB

```python
PropertyGraphIndex.from_existing(
    SimplePropertyGraphStore.from_persist_dir("./storage4"),
    vector_store=ChromaVectorStore(chroma_collection=collection)
)
```
